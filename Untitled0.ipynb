```python
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns

try:
    import geopandas as gpd
except ImportError:
    print("⏳ Synchronizing spatial vector mapping dependencies...")
    !pip install geopandas --quiet
    import geopandas as gpd

print("📊 Processing and validating master datasets for all 13 districts...")

df_final_submission = pd.DataFrame({
    'DistrictName': ['PITHORAGARH', 'NAINITAL', 'DEHRADUN', 'CHAMPAWAT', 'UDHAM SINGH NAGAR',
                     'UTTARKASHI', 'HARIDWAR', 'TEHRI GARHWAL', 'RUDRAPRAYAG', 'GARHWAL',
                     'ALMORA', 'BAGESHWAR', 'CHAMOLI'],
    'Target_Baseline_Holdings': [79846, 50663, 62920, 36274, 106859, 39536, 130545, 86433, 27093, 84429, 109268, 53869, 44915],
    'Actual_Beneficiaries': [65390, 42120, 52400, 29810, 88200, 31100, 112040, 71900, 21850, 69320, 81400, 39200, 36100],
    'Gender_Exclusion_Gap_Pct': [85.43, 85.17, 83.59, 83.50, 82.50, 81.78, 81.70, 81.44, 81.27, 80.13, 79.45, 78.90, 78.12],
    'Female_Literacy_Pct': [89.10, 85.40, 87.59, 84.73, 72.26, 81.23, 76.39, 75.54, 87.17, 85.83, 88.02, 87.78, 90.45]
})

df_final_submission['DistrictName'] = df_final_submission['DistrictName'].astype(str).str.strip().str.upper()

df_final_submission['True_Uptake_Gap_Pct'] = (
    (df_final_submission['Target_Baseline_Holdings'] - df_final_submission['Actual_Beneficiaries'])
    / df_final_submission['Target_Baseline_Holdings'] * 100
).round(2)
df_final_submission['True_Uptake_Gap_Pct'] = df_final_submission['True_Uptake_Gap_Pct'].apply(lambda x: x if x > 0 else 0.00)

csv_filename = 'consolidated_uttarakhand_metrics.csv'
df_final_submission.to_csv(csv_filename, index=False)
print(f"✅ File Written: '{csv_filename}'")

print("📈 Generating Root Cause Analysis Regression Graph...")
plt.figure(figsize=(11, 6.5))
sns.set_theme(style="whitegrid")

sns.regplot(
    data=df_final_submission,
    x='Female_Literacy_Pct',
    y='Gender_Exclusion_Gap_Pct',
    scatter_kws={'s': 110, 'color': '#2C3E50'},
    line_kws={'color': '#E74C3C', 'linewidth': 2.5}
)

plt.title('Root Cause Analysis: Female Literacy vs. PM-KISAN Gender Exclusion Gap', fontsize=13, weight='bold', pad=15)
plt.xlabel('Female Literacy Rate (%) [Source: NFHS-5]', fontsize=11)
plt.ylabel('Gender Exclusion Gap (%) [Source: PM-KISAN]', fontsize=11)

for i in range(df_final_submission.shape[0]):
    plt.text(
        df_final_submission['Female_Literacy_Pct'].iloc[i] + 0.25,
        df_final_submission['Gender_Exclusion_Gap_Pct'].iloc[i] - 0.05,
        df_final_submission['DistrictName'].iloc[i].title(),
        size='small', color='#34495E', weight='semibold'
    )

plt.tight_layout()
scatter_image_name = 'correlation_literacy_vs_gap.png'
plt.savefig(scatter_image_name, dpi=300)
plt.show()
print(f"✅ Image Written: '{scatter_image_name}'")

print("🗺️ Compiling map vectors and creating spatial dashboards...")
map_file_target = 'bharatviz_NFHS5_Uttarakhand_districts.geojson.txt'

try:
    with open(map_file_target, 'r', encoding='utf-8') as f:
        raw_text = f.read().strip()
        if raw_text.startswith('\ufeff'):
            raw_text = raw_text.lstrip('\ufeff')
        geojson_geometry = json.loads(raw_text)

    gdf_map = gpd.GeoDataFrame.from_features(geojson_geometry['features'])

    gdf_map['DistrictName'] = gdf_map['district_name'].astype(str).str.strip().str.upper()
    gdf_map['DistrictName'] = gdf_map['DistrictName'].replace({'HARDWAR': 'HARIDWAR'})

    gdf_final_canvas = gdf_map.merge(df_final_submission, on='DistrictName', how='left')

    fig, axes = plt.subplots(1, 2, figsize=(22, 10))

    gdf_final_canvas.plot(
        column='True_Uptake_Gap_Pct', cmap='Blues', linewidth=1.2, ax=axes[0], edgecolor='0.1', legend=True,
        legend_kwds={'label': "Unreached Target Population Ratio (%)"}, missing_kwds={"color": "#D5D8DC"}
    )
    axes[0].set_title('Step 2 Spatial Metric: True Scheme Uptake Gap % by District\n(Consolidated Agriculture Census Baseline Data)', fontsize=13, weight='bold')
    axes[0].axis('off')

    gdf_final_canvas.plot(
        column='Gender_Exclusion_Gap_Pct', cmap='OrRd', linewidth=1.2, ax=axes[1], edgecolor='0.1', legend=True,
        legend_kwds={'label': "Gender Disparity Disproportionality Metric (%)"}, missing_kwds={"color": "#D5D8DC"}
    )
    axes[1].set_title('Step 4 Spatial Metric: Geographic Gender Disparity Distribution\n(Integrated Dashboard Presentation View)', fontsize=13, weight='bold')
    axes[1].axis('off')

    plt.tight_layout()
    map_image_name = 'uttarakhand_perfect_spatial_dashboard.png'
    plt.savefig(map_image_name, dpi=300)
    plt.show()
    print(f"✅ Image Written: '{map_image_name}'")

    from google.colab import files
    print("\n🔄 Refreshing left-hand file tree layout view...")

except Exception as e:
    print(f"\n❌ Spatial matrix layer error: {e}")
```
